```python

#Find dice coefficient
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
total=0
np.set_printoptions(threshold=np.inf)
for i in range(51):
    
    
    
    
    img1=plt.imread('data/resize/'+str(i)+'.png')
    img1=np.asarray(img1)
    #print(img1)
    img1=np.where(img1<0.9,0,1)
    #img1=img1.astype(np.bool)
    #print(img1)
    #print(img1.shape)
  
    img2=plt.imread('data/segment_binary/'+str(i)+'.png')
    img2=np.asarray(img2)
    img2=np.where(img2<0.9,0,1)
    #print(img2)
    #img2=img2.astype(np.bool)
    #print(img2)
    #print(img2.shape)


    intersection = np.logical_and(img1, img2)
    val= 2. * intersection.sum() / (img1.sum() + img2.sum())
    print(i,val)
    total=total+val
    
print("Dice coeff=",(total/51))
    

```

```python
#Similarity index
from skimage.measure import compare_ssim

import cv2
total=0
np.set_printoptions(threshold=np.inf)
for i in range(51):
    imgx='data/resize/'+str(i)+'.png'
    imgy='data/segment_binary/'+str(i)+'.png'
    
    
    
    img1=cv2.imread(imgx,0)
    img2=cv2.imread(imgy,0)
    (val, diff) = compare_ssim(img1, img2, full=True)
    print(i,val)
    total=total+val
    
print("Similarity=",(total/51))
    

```

```python

#Find jaccard coefficient
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
total=0
np.set_printoptions(threshold=np.inf)
for i in range(51):
    
    
    
    
    img1=plt.imread('data/resize/'+str(i)+'.png')
    img1=np.asarray(img1)
    #print(img1)
    img1=np.where(img1<0.6,0,1)
    #img1=img1.astype(np.bool)
    #print(img1)
    #print(img1.shape)
  
    img2=plt.imread('data/segment_binary/'+str(i)+'.png')
    img2=np.asarray(img2)
    img1=np.where(img1<0.6,0,1)
    #print(img2)
    #img2=img2.astype(np.bool)
    #print(img2)
    #print(img2.shape)


    intersection = np.logical_and(img1, img2)
    union=np.logical_or(img1,img2)
    val= intersection.sum() / float(union.sum())
    total=total+val
    print(val)
    
print("Jaccard coeff=",(total/51))
    

```

```python
#Note that this is the non-black pixel area. Some of your rectangles have black inside them. So this is not the area 
#of the rectangle.You would have ensure that your image had no black pixels before isolating the rectangles to get 
#the area of the rectangles.

import cv2
import numpy as np
import imageio
total1=0
total2=0
for i in range(51):
    img1 = cv2.imread('data/resize/'+str(i)+'.png')
    height1 = img1.shape[0]
    width1 = img1.shape[1]
    #gray1 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
    ret1, thresh1 = cv2.threshold(img1,0,255,cv2.THRESH_BINARY)
    #imageio.imwrite('example.png',thresh1)
    ave1 = cv2.mean(thresh1)[0]/255
    area1 = ave1*height1*width1
    
    #########################for image 2#########################
    img2 = cv2.imread('data/segment_binary/'+str(i)+'.png')
    height2 = img2.shape[0]
    width2 = img2.shape[1]
    #gray2 = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
    ret2, thresh2 = cv2.threshold(img2,0,255,cv2.THRESH_BINARY)
    #imageio.imwrite('example_op.png',thresh2)
    ave2 = cv2.mean(thresh2)[0]/255
    area2 = ave2*height2*width2
    
    ###############################calculate the area#################################
    x=(area1*area2)
    y=max(area1,area2)
    ans=x/y
    print("Image[",str(i),"]:",ans)
    total1=total1+ans

print("FINAL AVERAGE BASED ON MANUAL COUNTING OF NON ZERO PIXELS",total1/51)
total_pixel=height1*width1
per_of_white_pixels=total1/total_pixel
print("SIMILARITY",per_of_white_pixels)
```
